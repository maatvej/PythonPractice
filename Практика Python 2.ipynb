### Задание №1

#### Найти пересечение без дубликатов

**Компания:** «Авито»

**Уровень:** Лёгкий

___

Пользователь с клавиатуры (через функцию `input()`) вводит через запятую и пробел две последовательности чисел (некоторые идентификаторы товаров/услуг, просмотренных клиентом 'A' и клиентом 'B' на платформе «Авито»).

Необходимо конвертировать каждый ввод пользователя в отдельный список, состоящий из целых чисел, а также составить новый список, состоящий из элементов пересечения двух исходных. Дубликаты необходимо удалить, а результирующий список — вывести на экран.

**Важно!** Если пересечение пустое, то на экран должен выводиться пустой список.

Если ввод пользователя невозможно конвертировать в список из целых чисел, возникает ошибка. Необходимо отлавливать эту ошибку и выводить на экран фразу `Некорректный ввод`.

Для поиска пересечений в списках напишите функцию `pure_intersection()`, которая будет возвращать список с уникальными элементами пересечения.


**Пример №1:**
```
Ввод:
108, 138, 42, 52, 14
109, 13, 52, 32, 42, 14, 109

Вывод:
[42, 52, 14]

```

**Пример №2:**
```
Ввод:
108, 138, 42, 52, 14
189, 202, 249, 24, 15, 3

Вывод:
[]

```

**Пример №3:**
```
Ввод:
108, 138, 42, 52, четырнадцать
189, 202, 249, 24, 15, 3

Вывод:
Некорректный ввод

```

**Подсказка:**

Для определения пересечения преобразуйте списки во множества и воспользуйтесь методом `intersection()`.

**Критерии оценивания:**

- Задача решена корректно, выполнены все условия (**2 балла**):
    - каждый ввод пользователя конвертируется в список из идентификаторов;
    - функция для определения пересечения корректно работает для любого ввода пользователя, обозначенного в условиях задачи;
    - решение сопровождено комментариями.
- Задача решена корректно, однако не выполнено одно из условий (**1 балл**):
    - ввод пользователя не конвертируется в список;
    - функция для определения пересечения работает только для ограниченного множества входных данных в рамках условия задачи;
    - решение не сопровождено комментариями.
- Задача решена с серьёзными ошибками/не решена (**0 баллов**): представлен некорректный алгоритм определения пересечения элементов.

````python
# ваш код здесь
def pure_intersection(list_a, list_b):
    # Возвращает список уникальных элементов пересечения двух списков
    return list(set(list_a).intersection(set(list_b)))

try:
    # Получаем ввод от пользователя и разбиваем его на два списка
    input_a = input("Введите идентификаторы для клиента A: ").strip()
    input_b = input("Введите идентификаторы для клиента B: ").strip()
    
    # Конвертируем ввод в списки целых чисел
    list_a = [int(x) for x in input_a.split(', ')]
    list_b = [int(x) for x in input_b.split(', ')]
    
    # Находим пересечение и выводим результат
    result = pure_intersection(list_a, list_b)
    print(result)
except ValueError:
    # Обрабатываем ошибку конвертации
    print("Некорректный ввод")
````

Комментарии к коду:
1. Функция `pure_intersection` принимает два списка и возвращает список уникальных элементов их пересечения.
2. Используем `input()` для получения данных от пользователя, затем разбиваем строку на элементы, используя `split(', ')`.
3. Преобразуем элементы в целые числа с помощью `int()`. Если ввод содержит некорректные данные, возникает `ValueError`, который мы обрабатываем, выводя сообщение "Некорректный ввод".
4. Если ввод корректен, выводим список пересечения.

## Задание №2

#### Найти минимум и максимум

**Компания:** «Авито»

**Уровень:** Лёгкий

___

Пользователь подаёт на вход программе действительные числа через пробел с клавиатуры (через функцию `input()`).

Он привык пользоваться Excel, поэтому отделяет дробную часть числа запятой, а не точкой. Кроме того, он может случайно ввести число «словами» (например, не "5", а «пять») — в этом случае слово надо игнорировать.

**Ваша задача** — вывести минимум и максимум последовательности. Гарантируется, что пользователем будет введено хотя бы одно число.

Для нахождения минимума и максимума напишите функцию `find_min_max`, которая будет возвращать два числа.

**Пример:**

```
Ввод:
-2,56 25,002 восемь 19,13 13 -37,5

Вывод:
Minimum: -37.5
Maximim: 25.002
```



**Подсказка №1:**

Для проверки того, состоит ли строка `s` только из букв, можно воспользоваться методом `s.isalpha()`.

**Подсказка №2:**

Для поиска минимума и максимума в списке можно воспользоваться встроенными функциями `min()` и `max()`.

**Критерии оценивания:**

- Задача решена корректно, учтены все условия (**2 балла**):
    - программа корректно обрабатывает ввод пользователя и работает на любых входных данных;
    - программа корректно обрабатывает присутствие слов во вводе пользователя;
    - программа корректно рассчитывает минимум и максимум числовой последовательности;
    - решение сопровождено комментариями.
- Задача решена корректно, но не выполнено одно из условий (**1 балл**):
    - программа некорректно обрабатывает ввод пользователя и работает только на конкретном примере входных данных;
    - программа некорректно обрабатывает присутствие слов во вводе пользователя, что приводит к ошибкам при выполнении программы;
    - решение не сопровождено комментариями.
- Задача решена с серьёзными ошибками/не решена (**0 баллов**): программа некорректно рассчитывает минимум и максимум числовой последовательности.

````python
def find_min_max(numbers):
    # Возвращает минимум и максимум из списка чисел
    return min(numbers), max(numbers)

# Получаем ввод от пользователя
user_input = input("Введите числа: ").strip()

# Разбиваем ввод на элементы
elements = user_input.split()

# Инициализируем список для хранения чисел
numbers = []

for element in elements:
    # Пропускаем элементы, которые являются словами
    if element.isalpha():
        continue
    # Заменяем запятую на точку и пытаемся преобразовать в число
    try:
        number = float(element.replace(',', '.'))
        numbers.append(number)
    except ValueError:
        # Игнорируем элементы, которые не удается преобразовать в число
        continue

# Находим минимум и максимум
minimum, maximum = find_min_max(numbers)

# Выводим результат
print(f"Minimum: {minimum}")
print(f"Maximum: {maximum}")
````

Комментарии к коду:
1. Функция `find_min_max` принимает список чисел и возвращает кортеж из минимального и максимального значений.
2. Пользовательский ввод разбивается на элементы с помощью `split()`.
3. Каждый элемент проверяется: если это слово (`isalpha()`), оно игнорируется.
4. Если элемент содержит запятую, она заменяется на точку, и элемент преобразуется в число `float()`.
5. Если преобразование не удается, элемент игнорируется.
6. После обработки всех элементов вычисляются и выводятся минимум и максимум.

## Задание №3

#### Найти медиану

**Компания:** «Авито»

**Уровень:** Лёгкий

---
Медиана, как и среднее, является одной из важнейших характеристик для числовых данных в статистике.

**Медиана** — это число, которое находится в середине упорядоченной по возрастанию числовой последовательности.

Например, в последовательности чисел `1, 5, 2, 3, 6` медианой является число `3`, так как если упорядочить числа по возрастанию (`1, 2, 3, 5, 6`), центральным элементом в последовательности будет именно `3`.

**Примечание.** *Если в выборке чётное число элементов, то за медиану принимается среднее между центральными элементами (после упорядочивания).*

Например, в последовательности чисел `100, 5, 2, 4, 3, 6` медианой является число `4.5`, так как, если упорядочить числа по возрастанию (`2, 3, 4, 5, 6, 100`), то центральным элементом в последовательности будет среднее арифметическое между 4 и 5 ($\frac{4 + 5} {2} = 4.5$).

---
Пользователь с клавиатуры (через функцию `input()`) вводит через запятую и пробел неупорядоченную последовательность чисел.

Необходимо конвертировать ввод пользователя в список действительных чисел и вывести на экран медиану этой числовой последовательности.

Для поиска медианы в списке напишите функцию `find_median()`, которая будет возвращать одно число — медианное значение. Результат работы функции нужно вывести на экран.

Если переданная в функцию последовательность окажется пустой (пользователь ничего не ввёл) или пользователь ввёл числа некорректно (ввод нельзя преобразовать в числа), на экран должна быть выведена фраза `Некорректный ввод`.

**Пример №1:**
```
Ввод:
1, 5, 2, 3, 6

Вывод:
Median: 3.0
```
```
Ввод:
100, 5, 2, 4, 3, 6

Вывод:
Median: 4.5
```
```
Ввод:

Вывод:
Некорректный ввод
```
```
Ввод:
десять, 10, пять, 7, семь

Вывод:
Некорректный ввод
```


**Критерии оценивания:**

- Задача решена корректно, выполнены все условия (**2 балла**):
    - ввод пользователя конвертируется в список из чисел;
    - функция для расчёта медианы работает корректно для списка чётной и нечётной длины;
    - в случае некорректного ввода функция возвращает `None`;
    - решение сопровождено комментариями.
- Задача решена корректно, однако не выполнено одно из условий (**1 балл**):
    - ввод пользователя не конвертируется в список;
    - функция для расчёта медианы некорректно работает для списка чётной или нечётной длины;
    - в случае некорректного ввода функция отрабатывает с ошибкой;
    - решение не сопровождено комментариями.
- Задача решена с серьёзными ошибками/не решена (**0 баллов**): представлен некорректный алгоритм расчёта медианы для обоих случаев, результат работы программы зависит от вводимой последовательности чисел и т. д.

````python
def find_median(numbers):
    if not numbers:
        return None
    
    # Сортировка списка чисел для нахождения медианы
    numbers.sort()
    n = len(numbers)
    mid = n // 2
    
    # Если в списке четное количество элементов, медиана представляет собой среднее значение двух средних чисел
    if n % 2 == 0:
        return (numbers[mid - 1] + numbers[mid]) / 2
    else:
        # Если в списке нечетное количество элементов, медианой является среднее число
        return numbers[mid]

def main():
    user_input = input("Введите числа через запятую и пробел: ")
    
    try:
        # Разделение входной строки запятыми и преобразование каждой части в число с плавающей точкой
        numbers = [float(num.strip()) for num in user_input.split(',')]
        if not numbers:
            raise ValueError
    except ValueError:
        print("Некорректный ввод. Пожалуйста, введите числа, разделенные запятыми.")
        return
    
    # Вычисление медианы с помощью функции find_median
    median = find_median(numbers)
    if median is not None:
        print(f"Median: {median}")
    else:
        print("Некорректный ввод. Убедитесь, что вы ввели хотя бы одно число.")

main()

````

````python
def transform_string_to_integer(s):
    
    # Словарь для сопоставления частей слов с числовыми значениями
    number_word_dict = {
        "ты": 1000, "м": 1000000,
        "сто": 100, "двес": 200, "трис": 300, "четырес": 400, "пятьс": 500, "шестьс": 600, "семьс": 700, "восемьс": 800, "девятьс": 900,
        "одинн": 11, "двен": 12, "трин": 13, "четырн": 14, "пятн": 15, "шестн": 16, "семн": 17, "восемн": 18, "девятн": 19,
        "двад": 20, "трид": 30, "сор": 40, "пятьд": 50, "шестьд": 60, "семьд": 70, "восемьд": 80, "девяно": 90,
        "дес": 10, "н": 0, "о": 1, "дв": 2, "т": 3, "ч": 4, "п": 5, "ш": 6, "с": 7, "в": 8, "д": 9,
    }
    
    # Разбиваем строку на слова
    words = s.split()
    total = 0 # Общая сумма
    current = 0 # Текущее число, которое мы формируем

    # Проходим по каждому слову
    for word in words:
        # Проверяем, начинается ли слово с какой-либо из ключевых частей
        for key in number_word_dict:
            if word.startswith(key):
                value = number_word_dict[key]  # Получаем числовое значение
                if value == 1000 or value == 1000000: # Если это тысяча или миллион, умножаем текущее число и добавляем к общей сумме
                    current *= value
                    total += current
                    current = 0 # Сбрасываем текущее число
                else:
                    current += value # Добавляем значение к текущему числу
                break

    total += current # Добавляем оставшееся текущее число к общей сумме
    return total

input_string = input("Введите число словами: ")
print(transform_string_to_integer(input_string))
````

### Задание №5 (бонусная, не оценивается)

#### Разложить факториал на простые множители

**Компания:** «СБЕР»

**Уровень:** Сложный

___

Пользователь с клавиатуры (через функцию `input()`) вводит некоторое целое число `n`.

Напишите функцию `decompose_factorial()`, которая будет принимать на вход введённое пользователем число `n`, а затем вычислять факториал этого числа и выполнять разложение факториала на простые множители.

---

Напомним, что **факториал числа $n$ ($n!$)** — это произведение всех натуральных чисел от $1$ до $n$ включительно.

Например, факториал числа $3$ будет равен: $3! = 1 * 2 * 3 = 6$.

---

**Простое число** — число, которое делится (нацело) только на себя и на 1. Например, $13$ — простое число, так как оно делится только на себя (на $13$) и на $1$, а $14$ не является простым числом, так как оно делится на $7$ и на $2$.

Подробнее о простых числах можно почитать [здесь](https://dzen.ru/a/XqFMe_VWclT8NTss).

---
**Простые множители** — это простые числа (не считая 1), на которые число делится без остатка.

Например, для числа $10$ простыми множителями являются числа $5$ и $2$. Для числа $28$ множителями являются числа $7$ и $4$, но простыми множителями будут числа $7$ и $2$, так как $4$ — это $2 * 2$, а число $28$ представляется в виде $28 = 7 * 2^2$.

----

В результате своей работы функция должна возвращать строку вида (приведён пример для `n=12`):

`12! = 2^10 * 3^5 * 5^2 * 7 * 11`

Здесь число слева от ^ — множитель, число справа от ^ — его степень. При этом если множитель встречается один раз, знак степени не выводится. Вывод множителей должен производиться в порядке их возрастания.

**Пример №1:**

Вывод:
5! = 2^3 * 3 * 5
```

**Пример №2:**

```

**Пример №3:**
```

```

**Подсказка №2:**

Алгоритм нахождения простых множителей числа вы можете найти [здесь](https://foxford.ru/wiki/informatika/razlozhenie-chisla-na-mnozhiteli-v-python).

````python
from math import factorial

def prime_factors(n):
    factors = {}
    d = 2
    while d * d <= n:
        while (n % d) == 0:
            if d in factors:
                factors[d] += 1
            else:
                factors[d] = 1
            n //= d
        d += 1
    if n > 1:
        factors[n] = 1
    return factors

def decompose_factorial(n):
    prime_count = {}
    for i in range(2, n + 1):
        factors = prime_factors(i)
        for prime, count in factors.items():
            if prime in prime_count:
                prime_count[prime] += count
            else:
                prime_count[prime] = count
    
    result = f"{n}! = "
    result += " * ".join(
        f"{prime}^{count}" if count > 1 else f"{prime}"
        for prime, count in sorted(prime_count.items())
    )
    return result

input_number = int(input("Введите число: "))
print(decompose_factorial(input_number))
````

**Подсказка:**

При решении задачи вам может понадобиться латинский алфавит. Вы можете получить его из встроенной в Python библиотеки `string`:

````python
import string
alphabet = string.ascii_lowercase
print(alphabet)
````

````python
import string

def replace_duplicates(s):
    alphabet = string.ascii_lowercase
    while True:
        found = False
        for char in alphabet:
            if s.count(char) >= 2:
                found = True
                s = s.replace(char, '', 2)
                next_char = alphabet[(alphabet.index(char) + 1) % len(alphabet)]
                s += next_char
                break
        if not found:
            break
    return ''.join(set(s))

input_string = input("Введите строку: ")
print(replace_duplicates(input_string))
````
